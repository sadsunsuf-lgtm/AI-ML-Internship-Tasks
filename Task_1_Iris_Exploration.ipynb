Import Important Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore') # Keeps the output clean
%matplotlib inline
```

Loading Dataset

```python
# Load the iris dataset from seaborn's built-in library
df = sns.load_dataset('iris')
print("Dataset loaded successfully!")
```

Exploratory Data Analysis (EDA) of the Iris Dataset

Data Inspection

```python
# Displaying dimensions and column names
print(f"Dataset Shape: {df.shape}")
print("-" * 30)
print(f"Columns: {list(df.columns)}")
print("-" * 30)
df.head()
```

"The dataset contains 150 rows and 5 columns (4 features and 1 target variable)."

```python
# Checking for data types and null values
df.info()
```

"Initial inspection via .info() confirms there are no missing values, meaning the data is clean and ready for visualization."

```python
# Statistical overview
df.describe()
```

"Based on .describe(), we can see that the 'petal_length' has a high standard deviation, suggesting significant variation between the different iris species."

Visualizing Relationships

Scatter Plot

```python
# Setting the aesthetic style of the plots
sns.set_theme(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species', palette='viridis', s=70)

plt.title('Sepal Length vs Sepal Width (Grouped by Species)', fontsize=15)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Iris Species')
plt.show()
```

Observation: The scatter plot reveals a clear cluster for the Setosa species (purple/darker dots), showing it has smaller sepal length but larger sepal width compared to the others. Versicolor and Virginica have some overlap, suggesting they are more closely related in sepal dimensions.

Histograms

```python
# Plotting histograms for all numerical features
df.hist(edgecolor='black', linewidth=1.2, figsize=(12, 8), color='skyblue')
plt.suptitle('Distribution of Numerical Features', fontsize=16)
plt.show()
```

Box Plots

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, palette="Set3")
plt.title('Box Plot of Iris Features', fontsize=15)
plt.show()
```

```python
# Calculating correlation only for numeric columns
correlation_matrix = df.corr(numeric_only=True)
correlation_matrix
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Iris Features')
plt.show()
```
